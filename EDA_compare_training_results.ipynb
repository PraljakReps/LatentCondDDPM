```python
import pandas as pd
import matplotlib.pyplot as plt
```

## CIFAR10:

```python
cifar_vanDDPM = pd.read_csv('./output/vanillaDDPM/cifar10/model_train_history.csv')
cifar_latentDDPM = pd.read_csv('./output/latentDDPM/cifar10/model_train_history.csv')
cifar_SSlatentDDPM = pd.read_csv('./output/SSlatentDDPM/cifar10/model_train_history.csv')
```

```python
fig, axes = plt.subplots(1,4, dpi = 300, figsize = (14,3))


axes[0].plot(cifar_vanDDPM['train_MSE'], label = 'Vanilla-DDPM')
axes[0].plot(cifar_latentDDPM['train_mse_loss'], label = 'latent-DDPM')
axes[0].plot(cifar_SSlatentDDPM['train_mse_loss'], label = 'SS-latent-DDPM')

axes[1].plot(cifar_vanDDPM['train_MSE'])
axes[1].plot(cifar_latentDDPM['train_mse_loss'])
axes[1].plot(cifar_SSlatentDDPM['train_mse_loss'])

axes[2].plot(cifar_vanDDPM['test_MSE'], label = 'Vanilla-DDPM')
axes[2].plot(cifar_latentDDPM['test_mse_loss'], label = 'latent-DDPM')
axes[2].plot(cifar_SSlatentDDPM['test_mse_loss'], label = 'SS-latent-DDPM')

axes[3].plot(cifar_vanDDPM['test_MSE'])
axes[3].plot(cifar_latentDDPM['test_mse_loss'])
axes[3].plot(cifar_SSlatentDDPM['test_mse_loss'])

axes[0].legend()
axes[1].set_ylim([29, 35])
axes[3].set_ylim([29, 35])

axes[0].set_title('Train set')
axes[1].set_title('Train set')
axes[2].set_title('Test set')
axes[3].set_title('Test set')


axes[0].set_ylabel('MSE loss ($\epsilon - \epsilon_{\Theta}$)')


axes[0].set_xlabel('Epochs')
axes[1].set_xlabel('Epochs')
axes[2].set_xlabel('Epochs')
axes[3].set_xlabel('Epochs')


plt.savefig('./output/figures/cifar10_compare_DDPM_MSE.png', dpi = 300)
```
